🔹 What are Vectors?
A vector is simply a list of numbers (in 1D, 2D, or more dimensions).
Example:

1D Vector (Scalar): [3]

2D Vector: [2, 4]

3D Vector: [1, -3, 5]

🔹 Key Operations on Vectors:

Addition: Add corresponding elements.
Multiplication (Scalar & Dot Product):
Scalar Multiplication: Multiply each element by a scalar.
Dot Product: Multiply corresponding elements and sum them.

```python
import numpy as np
import matplotlib.pyplot as plt

#############################
# Operations using Python Lists
#############################
v1_list = [1, 2, 4]
v2_list = [3, 5, 6]

# List Addition results in concatenation
list_add = v1_list + v2_list
# List Multiplication results in repetition
list_mult = 3 * v1_list

print("Python Lists:")
print("v1:", v1_list)
print("v2:", v2_list)
print("List Addition (concatenation):", list_add)
print("List Multiplication (repetition):", list_mult)


#############################
# Operations using NumPy Arrays
#############################
v1 = np.array([1, 2, 4])
v2 = np.array([3, 5, 6])

# Vector Addition (element-wise)
vector_sum = v1 + v2
# Scalar Multiplication
scalar_mult = 3 * v1
# Dot Product Calculation
dot_product = np.dot(v1, v2)

print("\nNumPy Arrays:")
print("v1:", v1)
print("v2:", v2)
print("Vector Addition:", vector_sum)       # Expected: [4, 7, 10]
print("Scalar Multiplication:", scalar_mult) # Expected: [3, 6, 12]
print("Dot Product:", dot_product)           # Expected: 37


#############################
# Matrix-Style Visualization Function
#############################
def plot_matrix(ax, matrix, title):
    """
    Displays the given matrix (converted to at least 2D) as a heatmap,
    with each cell annotated by its value.
    """
    matrix = np.atleast_2d(matrix)  # Convert 1D vector to a 2D row vector
    cax = ax.matshow(matrix, cmap="Pastel1", alpha=0.75)
    for (i, j), val in np.ndenumerate(matrix):
        ax.text(j, i, str(val), ha='center', va='center', fontsize=12, color='black', fontweight='bold')
    ax.set_xticks(range(matrix.shape[1]))
    ax.set_yticks(range(matrix.shape[0]))
    ax.set_title(title, fontsize=12, pad=10)
    plt.colorbar(cax, ax=ax)
```

```python
#############################
# Visualization for Python List Operations
#############################
# For visualization, convert lists to NumPy arrays.
v1_list_arr = np.array(v1_list)
v2_list_arr = np.array(v2_list)
list_add_arr = np.array(list_add)
list_mult_arr = np.array(list_mult)

# Create subplots for list v1, v2, and addition (concatenation)
fig_list, axes_list = plt.subplots(1, 3, figsize=(15, 5))
plot_matrix(axes_list[0], v1_list_arr, "List v1")
plot_matrix(axes_list[1], v2_list_arr, "List v2")
plot_matrix(axes_list[2], list_add_arr, "List Addition (Concatenation)")
plt.tight_layout()
plt.show()

# Visualize list multiplication (repetition) in its own plot
plt.figure(figsize=(5, 2))
plot_matrix(plt.gca(), list_mult_arr, "List Multiplication (Repetition)")
plt.tight_layout()
plt.show()
```

```python
#############################
# Visualization for NumPy Array/ Vector Operations
#############################
fig_np, axes_np = plt.subplots(2, 3, figsize=(15, 10))
# Row 0: Show original vectors and their addition
plot_matrix(axes_np[0,0], v1, "Vector v1")
plot_matrix(axes_np[0,1], v2, "Vector v2")
plot_matrix(axes_np[0,2], vector_sum, "Vector Addition")

# Row 1: Show scalar multiplication and dot product (as 1x1 matrix)
plot_matrix(axes_np[1,0], scalar_mult, "Scalar Multiplication")
plot_matrix(axes_np[1,1], np.array([dot_product]), "Dot Product")
axes_np[1,2].axis('off')  # Hide the unused subplot

plt.tight_layout()
plt.show()
```

🔹 Why is the cross product important?

Used in Computer Graphics (Finding Normals to Surfaces)

```python
# Define two 3D vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Compute cross product
cross_product = np.cross(a, b)

print("Cross Product:", cross_product)  # Output: [-3, 6, -3]

VectorFigure = plt.figure(figsize=(10, 7))
ax = VectorFigure.add_subplot(111, projection='3d')

origin = np.array([0, 0, 0])

vectors = {
    "a " : (a, 'r'),
    "b ": (b, 'b'),
    "Cross Product a X b" : (cross_product, 'g')
}

for label, (vector, color) in vectors.items(): # .items() only used for dictionaries not lists or tuples
    ax.quiver(*origin, *vector, color=color, label=label, arrow_length_ratio=0.1)

ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.set_zlabel("Z-Axis")
ax.set_title("Cross Product Visualization")

# Set axis limits dynamically
ax.set_xlim([0, max(abs(cross_product[0]), abs(b[0])) + 1])
ax.set_ylim([0, max(abs(cross_product[1]), abs(b[1])) + 1])
ax.set_zlim([0, max(abs(cross_product[2]), abs(b[2])) + 1])

ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# Matrix-Style Visualization
# ---------------------------
# Function to display an array as a matrix with text annotations
def plot_matrix(ax, matrix, title):
    # Convert 1D vectors to a 2D array (1xN) so that matshow can display them
    matrix = np.atleast_2d(matrix)
    cax = ax.matshow(matrix, cmap="Pastel1", alpha=0.75)
    # Annotate each cell with its corresponding value
    for (i, j), val in np.ndenumerate(matrix):
        ax.text(j, i, f'{val}', ha='center', va='center', fontsize=12, color='black', fontweight='bold')
    # Set ticks corresponding to cell indices
    ax.set_xticks(range(matrix.shape[1]))
    ax.set_yticks(range(matrix.shape[0]))
    ax.set_title(title, fontsize=12, pad=10)
    # Optionally, add a colorbar for reference
    plt.colorbar(cax, ax=ax)

# Create a figure with subplots for the matrix visualization of the vectors
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
vectors_1d = {
    "Vector a": a,
    "Vector b": b,
    "Cross Product": cross_product
}

for ax, (title, vec) in zip(axes, vectors_1d.items()):
    plot_matrix(ax, vec, title)

plt.tight_layout()
plt.show()
```

🔹 Key Operations on Matrices:

Matrix Addition: Add corresponding elements.
Matrix Multiplication: Multiply rows of first matrix with columns of second.
Transpose: Flip the matrix along its diagonal.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

##  Element-wise Multiplication (Hadamard Product)
E_mul =  A * B

## matrics multiplication
matrix_mul_AB = np.dot(A, B)

print("A: ", A)
print ("B: ", B)
print("Element-wise Multiplication: ", E_mul)
print("Matrics multiplication: ", matrix_mul)


C = np.array([[2, 4, 6], [3, 5, 7]])
D = np.array([[1, 3, 5], [9, 8, 7]])

##  Element-wise Multiplication (Hadamard Product)
mult = C*D

## matrics multiplication
matrix_mul_CD = np.dot(C, D.T)  # or C @ D.T

print("C: ", C)
print ("D: ", D)
print("Element-wise Multiplication: ", mult)
print("Matrics multiplication: ", matrix_mul)

# Create figure with subplots for corrected visualization
fig, axes = plt.subplots(2, 6, figsize=(15, 8))

# Define a function to display a matrix with text annotations
def plot_matrix(ax, matrix, title):
    ax.matshow(matrix, cmap="Pastel1", alpha=0.75)
    for (i, j), val in np.ndenumerate(matrix):
        ax.text(j, i, f'{val}', ha='center', va='center', fontsize=12, color='black', fontweight='bold')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title, fontsize=12, pad=10)


# Plot each matrix with corrected results
plot_matrix(axes[0, 0], A, "Matrix A")
plot_matrix(axes[0, 1], B, "Matrix B")
plot_matrix(axes[0, 2], E_mul, "Element-wise Mul (A * B)")
plot_matrix(axes[0, 3], A, "Matrix A")
plot_matrix(axes[0, 4], B, "Matrix B")
plot_matrix(axes[0, 5], matrix_mul_AB, "Matrix Mul (A @ B)")

# # Hide the unused subplot in row 1, column 5
# axes[0, 4].axis("off")

plot_matrix(axes[1, 0], C, "Matrix C")
plot_matrix(axes[1, 1], D, "Matrix D")
plot_matrix(axes[1, 2], mult, "Element-wise Mul (C * D)")
plot_matrix(axes[1, 3], C, "Matrix C")
plot_matrix(axes[1, 4], D.T, "Matrix transpose D.T")
plot_matrix(axes[1, 5], matrix_mul_CD, "Matrix Mul (C @ D.T)")

# Adjust layout and show the corrected plot
plt.tight_layout()
plt.show()
```

🔹 What are Eigenvalues & Eigenvectors?
Eigenvalues & Eigenvectors help in dimensionality reduction (PCA, SVD) and are used in many ML algorithms.

Eigenvectors: Special vectors that only scale when multiplied by a matrix.
Eigenvalues: The scaling factors associated with each eigenvector.
Mathematically:

𝐴

A = Matrix

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[4, -2],
              [1,  1]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Create a figure for visualization
fig, ax = plt.subplots(figsize=(8, 8))

# Draw coordinate axes
ax.axhline(0, color='grey', lw=0.5)
ax.axvline(0, color='grey', lw=0.5)
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_aspect('equal')

# 1. Plot the unit circle (dashed black) representing all unit directions.
theta = np.linspace(0, 2*np.pi, 200)
x_unit = np.cos(theta)
y_unit = np.sin(theta)
ax.plot(x_unit, y_unit, 'k--', label="Unit Circle")

# 2. Plot the transformed unit circle (green) showing A acting on every unit vector.
unit_circle = np.array([x_unit, y_unit])
transformed_circle = A @ unit_circle
ax.plot(transformed_circle[0], transformed_circle[1], 'g-', label="Transformed Circle (Ellipse)")

colors = ['r', 'b']
for i in range(len(eigenvalues)):
    # Extract the i-th eigenvector (from the columns of eigenvectors)
    vec = eigenvectors[:, i]
    # Normalize to unit length to show its direction
    vec_unit = vec / np.linalg.norm(vec)
    # Scale the eigenvector by its eigenvalue
    scaled_vec = eigenvalues[i] * vec_unit

    # Plot the eigenvector unit direction as a dotted arrow using annotate.
    # We use a custom dash tuple (0, (1, 5)) to simulate a dotted line.
    ax.annotate("",
                xy=(vec_unit[0], vec_unit[1]),
                xytext=(0, 0),
                arrowprops=dict(arrowstyle="->", color=colors[i], linestyle=(0, (1, 5)), lw=2))

    # Plot the scaled eigenvector as a solid arrow using quiver.
    ax.quiver(0, 0, scaled_vec[0], scaled_vec[1],
              angles='xy', scale_units='xy', scale=1,
              color=colors[i],
              label=f"Eigenvalue {eigenvalues[i]:.2f}×Eigenvector {i+1}")

    # Annotate the tip of the scaled eigenvector with its eigenvalue.
    ax.text(scaled_vec[0], scaled_vec[1], f"λ={eigenvalues[i]:.2f}", fontsize=12,
            color=colors[i], ha='center', va='center')

# 4. Optional: Illustrate transformation on arbitrary vectors.
points = np.array([[2, 0], [0, 2], [2, 2], [-2, 1]]).T  # shape (2,4)
transformed_points = A @ points

# Plot the original arbitrary vectors (magenta circles)
ax.scatter(points[0], points[1], c='magenta', marker='o', label="Original vectors")
# Plot the transformed arbitrary vectors (cyan crosses)
ax.scatter(transformed_points[0], transformed_points[1], c='cyan', marker='x', label="Transformed vectors")
# Draw arrows connecting original and transformed vectors
for i in range(points.shape[1]):
    ax.arrow(points[0,i], points[1,i],
             transformed_points[0,i]-points[0,i],
             transformed_points[1,i]-points[1,i],
             head_width=0.3, head_length=0.3, fc='gray', ec='gray')

ax.set_title("Visualization of Eigenvectors and Eigenvalues")
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[4, -2],
              [1,  1]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Generate a grid of points in the plane
x = np.linspace(-5, 5, 11)
y = np.linspace(-5, 5, 11)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# Transform the grid by applying A
transformed_points = A @ points

# Create subplots for original and transformed grids
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the original grid
axes[0].scatter(points[0], points[1], color='blue', s=30, alpha=0.7)
axes[0].set_title("Original Grid")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].axhline(0, color='black', lw=0.5)
axes[0].axvline(0, color='black', lw=0.5)
axes[0].set_aspect('equal')

# Overlay the eigenvectors (scaled by their eigenvalues) on the original grid
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i]
    # Normalize to get direction only
    vec_unit = vec / np.linalg.norm(vec)
    # Scale by the eigenvalue to show the stretching
    scaled_vec = eigenvalues[i] * vec_unit
    axes[0].quiver(0, 0, scaled_vec[0], scaled_vec[1],
                   angles='xy', scale_units='xy', scale=1,
                   color='red', label=f"Eigen {i+1}: λ={eigenvalues[i]:.2f}")
axes[0].legend()

# Plot the transformed grid
axes[1].scatter(transformed_points[0], transformed_points[1], color='purple', s=30, alpha=0.7)
axes[1].set_title("Transformed Grid (by A)")
axes[1].set_xlabel("x'")
axes[1].set_ylabel("y'")
axes[1].axhline(0, color='black', lw=0.5)
axes[1].axvline(0, color='black', lw=0.5)
axes[1].set_aspect('equal')

# Overlay the eigenvectors on the transformed grid (their directions remain unchanged)
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i]
    vec_unit = vec / np.linalg.norm(vec)
    scaled_vec = eigenvalues[i] * vec_unit
    axes[1].quiver(0, 0, scaled_vec[0], scaled_vec[1],
                   angles='xy', scale_units='xy', scale=1,
                   color='red', label=f"Eigen {i+1}: λ={eigenvalues[i]:.2f}")
axes[1].legend()

plt.suptitle("Grid Transformation by Matrix A and Invariant Eigenvector Directions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # For displaying images in Colab

# Create a blank white image (600x600)
img_size = 600
img = np.ones((img_size, img_size, 3), dtype=np.uint8) * 255

# Helper function to convert mathematical (x, y) coordinates
# to image coordinates (origin at center, y-axis inverted)
def to_image_coords(x, y, img_width=img_size, img_height=img_size, scale=50):
    cx, cy = img_width // 2, img_height // 2
    return int(cx + x * scale), int(cy - y * scale)

# Draw coordinate axes (light gray)
cv2.line(img, (0, img_size//2), (img_size, img_size//2), (200, 200, 200), 1)
cv2.line(img, (img_size//2, 0), (img_size//2, img_size), (200, 200, 200), 1)

# Draw the unit circle (approximate with many points)
theta = np.linspace(0, 2*np.pi, 200)
unit_circle_pts = []
for t in theta:
    x, y = np.cos(t), np.sin(t)
    unit_circle_pts.append(to_image_coords(x, y))
unit_circle_pts = np.array(unit_circle_pts, np.int32).reshape((-1, 1, 2))
cv2.polylines(img, [unit_circle_pts], isClosed=True, color=(0, 0, 0), thickness=1)

# Draw the transformed unit circle (ellipse) by applying A to each point
transformed_pts = []
for t in theta:
    vec = np.array([np.cos(t), np.sin(t)])
    x, y = A @ vec
    transformed_pts.append(to_image_coords(x, y))
transformed_pts = np.array(transformed_pts, np.int32).reshape((-1, 1, 2))
cv2.polylines(img, [transformed_pts], isClosed=True, color=(0, 255, 0), thickness=1)

# Define colors for the eigenvectors
colors = [(0, 0, 255), (255, 0, 0)]  # red and blue

# Draw eigenvectors:
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i]
    vec_unit = vec / np.linalg.norm(vec)
    scaled_vec = eigenvalues[i] * vec_unit

    # Compute start and end points for drawing (in image coordinates)
    origin = to_image_coords(0, 0)
    end_unit = to_image_coords(vec_unit[0], vec_unit[1])
    end_scaled = to_image_coords(scaled_vec[0], scaled_vec[1])

    # Draw dashed line for unit eigenvector by simulating dashes
    num_segments = 15
    for j in range(num_segments):
        t0 = j / num_segments
        t1 = (j + 0.5) / num_segments
        x0, y0 = vec_unit[0] * t0, vec_unit[1] * t0
        x1, y1 = vec_unit[0] * t1, vec_unit[1] * t1
        pt1 = to_image_coords(x0, y0)
        pt2 = to_image_coords(x1, y1)
        cv2.line(img, pt1, pt2, colors[i], 2)

    # Draw solid arrow for the scaled eigenvector
    cv2.arrowedLine(img, origin, end_scaled, colors[i], 2, tipLength=0.1)
    # Annotate with eigenvalue near the tip
    cv2.putText(img, f"λ={eigenvalues[i]:.2f}", end_scaled,
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors[i], 1, cv2.LINE_AA)

# Show the final image using cv2_imshow (compatible with Colab)
cv2_imshow(img)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[4, -2],
              [1,  1]])

# Compute eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eig(A)

# Define an arbitrary (normal) vector v
v = np.array([1, 2])
# Transform the vector: v_trans = A*v
v_trans = A @ v

# Decompose v in the eigenbasis: solve eigvecs * c = v for coefficients c
c = np.linalg.solve(eigvecs, v)  # c such that v = c1*eigvec1 + c2*eigvec2

# Reconstruct v from its eigen components (for verification)
v_recon = c[0]*eigvecs[:,0] + c[1]*eigvecs[:,1]
# And note that after transformation, each component is scaled:
v_trans_decomp = c[0]*eigvals[0]*eigvecs[:,0] + c[1]*eigvals[1]*eigvecs[:,1]

# Prepare two subplots: one for the original vector decomposition, one for the transformed vector.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))

# --- Panel 1: Original Vector and its Eigenvector Components ---
for ax in [ax1, ax2]:
    ax.axhline(0, color='black', lw=0.5)
    ax.axvline(0, color='black', lw=0.5)
    ax.set_aspect('equal', 'box')
    ax.grid(True, linestyle='--', alpha=0.5)

ax1.set_xlim(-5, 5)
ax1.set_ylim(-5, 5)
ax1.set_title("Original Vector Decomposition")

# Colors for eigenvectors/components
colors = ['r', 'b']

# Plot eigenvectors (normalized) as dashed arrows
for i in range(2):
    ev = eigvecs[:, i]
    ev_unit = ev / np.linalg.norm(ev)
    ax1.arrow(0, 0, 3*ev_unit[0], 3*ev_unit[1],
              head_width=0.2, head_length=0.2,
              fc=colors[i], ec=colors[i], linestyle='dashed',
              label=f"Eigenvector {i+1}\nλ={eigvals[i]:.2f}")

# Plot the original vector v in green
ax1.arrow(0, 0, v[0], v[1], head_width=0.2, head_length=0.2,
          fc='g', ec='g', lw=2, label=f"v = {v}")
ax1.text(v[0]*1.05, v[1]*1.05, f"{v}", color='g', fontsize=12)

# Plot the eigen-components of v
comp1 = c[0]*eigvecs[:,0]
comp2 = c[1]*eigvecs[:,1]
ax1.arrow(0, 0, comp1[0], comp1[1],
          head_width=0.15, head_length=0.15, fc='m', ec='m',
          label=f"Component 1 = {comp1.round(2)}")
ax1.arrow(comp1[0], comp1[1], comp2[0], comp2[1],
          head_width=0.15, head_length=0.15, fc='c', ec='c',
          label=f"Component 2 = {comp2.round(2)}")
ax1.legend(loc='upper left', fontsize=10)

# --- Panel 2: Transformed Vector and its Scaled Eigenvector Components ---
ax2.set_xlim(-30, 30)
ax2.set_ylim(-30, 30)
ax2.set_title("Transformed Vector (A*v) Decomposition")

# Plot eigenvectors for reference (scaled for visibility)
for i in range(2):
    ev = eigvecs[:, i]
    ev_unit = ev / np.linalg.norm(ev)
    ax2.arrow(0, 0, 5*ev_unit[0], 5*ev_unit[1],
              head_width=0.5, head_length=0.5,
              fc=colors[i], ec=colors[i], linestyle='dashed',
              label=f"Eigenvector {i+1}\nλ={eigvals[i]:.2f}")

# Plot the transformed vector v_trans in green (thicker arrow)
ax2.arrow(0, 0, v_trans[0], v_trans[1], head_width=0.5, head_length=0.5,
          fc='g', ec='g', lw=2, label=f"A*v = {v_trans.round(2)}")
ax2.text(v_trans[0]*1.05, v_trans[1]*1.05, f"{v_trans.round(2)}", color='g', fontsize=12)

# Plot the transformed eigen-components
comp1_trans = c[0]*eigvals[0]*eigvecs[:,0]
comp2_trans = c[1]*eigvals[1]*eigvecs[:,1]
ax2.arrow(0, 0, comp1_trans[0], comp1_trans[1],
          head_width=0.4, head_length=0.4, fc='m', ec='m',
          label=f"Component 1 Transformed = {comp1_trans.round(2)}")
ax2.arrow(0, 0, comp1_trans[0]+comp2_trans[0], comp1_trans[1]+comp2_trans[1],
          head_width=0.4, head_length=0.4, fc='c', ec='c',
          label=f"Component 2 Transformed = {comp2_trans.round(2)}")
ax2.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()
```
